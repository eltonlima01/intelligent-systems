1. Calcule a correlacão entre as variáveis independentes (AT, V, AP, RH) e a variável dependente (PE). Quais variáveis parecem ter maior influência na producão de energia?

```python
import pandas

dataset = pandas.read_csv("./datasets/Power Plant Data.csv")
dataset

x = dataset[["AT", "V", "AP", "RH"]]
y = dataset["PE"]

corr = x.corrwith(y)
print(corr)
```

Pela interpretação dos valores, as variáveis independentes AT e V apresentam uma correlação muito forte (de magnitudes 0.948128 e 0.869780, respectivamente) com a variável dependente PE, mas inversamente proporcionais (sinal negativo).

Por outro lado, RH e AP apresentam correlações moderada e forte (magnitudes 0.389794 e 0.518429), ambas diretamente proporcionais (sinal positivo).
